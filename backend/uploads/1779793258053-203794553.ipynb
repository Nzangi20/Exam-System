## Advanced Data Types
### 1. Collections
Python’s `collections` module provides specialized container datatypes beyond built-in types like list, tuple, dict, and set.

- `Counter`: For counting hashable objects
- `defaultdict`: Like a dictionary, but provides a default value for the key that does not exist
- `OrderedDict`: A dict that remembers insertion order (now built-in in modern Python)
- `namedtuple`: Factory function for creating tuple subclasses with named fields
- `deque`: A list-like container with fast appends and pops on either end

The code below uses `Counter` to count how many times each fruit appears in the list.
It then prints a dictionary-like output showing the frequency of each fruit.

````python
from collections import Counter, defaultdict, namedtuple, deque

# Counter example
cnt = Counter(['apple', 'orange', 'apple', 'pear', 'orange', 'banana'])
print(cnt)
````

This code uses a `defaultdict` with integers to automatically initialize missing keys to zero and count how many times each item appears.
It prints a dictionary showing the frequency of each fruit without needing to check if a key already exists.


````python
# defaultdict example
d = defaultdict(int)
for item in ['apple', 'orange', 'apple']:
    d[item] += 1
print(d)
````

This code creates a `namedtuple` called `Point` with fields `x` and `y`, allowing access to values by name instead of index.
It then creates a point with coordinates `(1, 2)` and prints the `x` and `y` values.

````python
# namedtuple example
Point = namedtuple('Point', ['x', 'y'])
p = Point(1, 2)
print(p.x, p.y)
````

This code creates a `deque` with initial elements and adds an item to both the left and right ends.
It then prints the updated deque, showing efficient insertion at both ends.

````python
# deque example
q = deque([1, 2, 3])
q.appendleft(0)
q.append(4)
print(q)
````

## Functional Programming
Python supports functional programming paradigms using built-in functions like `map`, `filter`, `reduce`, and anonymous functions (`lambda`).

* **Map:** This code uses `map()` to apply a function that squares each number in the list, producing a new list of squared values.
* **Filter:** This code uses `filter()` to select only the even numbers from the list, returning a list that meets the condition.
* **Reduce:** This code uses `reduce()` to multiply all the numbers in the list together, resulting in a single value that is the product of all elements.


````python
from functools import reduce

# map example
nums = [1, 2, 3, 4, 5, 6, 7, 8, 10]
squares = list(map(lambda x: x**2, nums))
print("Squares:", squares)

# filter example
even = list(filter(lambda x: x % 2 == 0, nums))
print("Even Numbers:", even)

# reduce example
product = reduce(lambda x, y: x * y, nums)
print("Product:", product)
````

## Comprehensions
Comprehensions provide a concise way to create lists, dictionaries, and sets from iterables.

* **List comprehension:** This code creates a list containing the squares of numbers from 0 to 9 in a concise and readable way.
* **Dictionary comprehension:** This code creates a dictionary where each key is a number and its value is the square of that number.
* **Set comprehension:** This code creates a set of squared values, automatically removing any duplicate results.


````python
# List comprehension
squares = [x**2 for x in range(10)]
print(squares)

# Dictionary comprehension
square_dict = {x: x**2 for x in range(5)}
print(square_dict)

# Set comprehension
unique_squares = {x**2 for x in [1, 2, 3]}
print(unique_squares)
````

## File I/O
Python makes it easy to read and write files such as text, CSV, and JSON.

* **Text file I/O:** This code writes a short message to a text file and then reads the file contents back and prints them.


````python
# Writing and reading a text file
with open("example.txt", "w") as f:
    f.write("Hello, world!\nPython file I/O.")

with open("example.txt", "r") as f:
    print(f.read())
````

* **JSON handling:** This code saves a Python dictionary to a JSON file and then loads it back into Python, preserving the data structure.

````python
# JSON
import json
data = {'name': 'Alice', 'age': 25, 'city': 'Nairobi'}
with open('data.json', 'w') as f:
    json.dump(data, f)

with open('data.json', 'r') as f:
    loaded = json.load(f)
    print(loaded)
````

## Practical Lab: Data Cleaning Without Libraries
**Objective:** Use core Python to clean a mock dataset loaded from a text file, including stripping whitespace, filtering missing entries, and computing basic stats.

````python
# Sample data (normally loaded from file)
data = [
    " Name , Age , Score ",
    " Alice , 24 , 88 ",
    " Bob , , 75 ",
    " Charlie , 22 , 90",
    " Dana , 27 , 85 "
]
data
````

```python
header = [col.strip() for col in data[0].split(',')]
```

* Take the first row of `data` (the header row).
* Split the string by commas to separate column names.
* Use `.strip()` to remove extra spaces from each column name.
* Store the clean column names in the list `header`.

```python
records = []
```

* Create an empty list `records` to store the cleaned data rows as dictionaries.

```python
for row in data[1:]:
```

* Loop over all rows in `data` **except the header** (i.e., from the second row onward).

```python
fields = [field.strip() for field in row.split(',')]
```

* Split the current row string by commas to get individual field values.
* Strip leading and trailing spaces from each field.
* Store the cleaned values in the list `fields`.

```python
if all(fields):
```

* Check if **all fields have a value** (i.e., no empty or missing data).
* Only proceed if the row is complete.

```python
records.append(dict(zip(header, fields)))
```

* Pair each field value with its corresponding column name using `zip(header, fields)`.
* Convert the zipped pairs into a dictionary.
* Append the dictionary to the `records` list.

```python
print("Cleaned Records:")
for rec in records:
    print(rec)
```

* Print the title `"Cleaned Records:"`.
* Loop through each dictionary in `records` and print it to show the cleaned data.

````python
# Clean and process data
header = [col.strip() for col in data[0].split(',')]
records = []
for row in data[1:]:
    fields = [field.strip() for field in row.split(',')]
    if all(fields):
        records.append(dict(zip(header, fields)))

print("Cleaned Records:")
for rec in records:
    print(rec)
````

Finally, the program extracts the Score values from the cleaned records, converts them to integers, computes the average score, and prints the result.

````python
# Compute average score
scores = [int(rec['Score']) for rec in records]
avg_score = sum(scores) / len(scores)
print("Average Score:", avg_score)
````
